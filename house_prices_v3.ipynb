```python
# This Python 3 environment comes with many helpful analytics libraries installed
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import torch
from torch import nn
from torch.utils.data import DataLoader, TensorDataset

from sklearn.ensemble import RandomForestRegressor
# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
X_train = pd.read_csv('/kaggle/input/idkidk/train.csv')
X_test = pd.read_csv('/kaggle/input/idkidk/test.csv')
y_train = X_train['SalePrice']
X_train = X_train.fillna(0)
X_test = X_test.fillna(0)

X_train = X_train.drop(columns=['SalePrice'])

cat_mask = (X_train.dtypes == "object").values
X_train_cat = X_train.loc[:, cat_mask]
X_test_cat = X_test.loc[:, cat_mask]
X_train_no = X_train.loc[:, ~cat_mask]
X_test_no = X_test.loc[:, ~cat_mask]
comb = pd.concat([X_train_cat, X_test_cat], axis=0)
comb_dummies = pd.get_dummies(comb, drop_first=True)

train_size = len(X_train_cat)
train = comb_dummies[:train_size]  
test = comb_dummies[train_size:]
train = train.replace({True: 1, False: 0})
test = test.replace({True: 1, False: 0})
X_train0 = pd.concat([train.reset_index(drop=True), X_train_no.reset_index(drop=True)], axis=1)
X_test0 = pd.concat([test.reset_index(drop=True), X_test_no.reset_index(drop=True)], axis=1)
```

```python
model = RandomForestRegressor(random_state=42, n_estimators=300)
model.fit(X_train0, y_train)

# Predict the target for the test set
y_pred = model.predict(X_test0)

output = pd.DataFrame({'Id': X_test.Id, 'SalePrice': y_pred})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
